```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('goal14.kobe.csv')
data.head()
```

```python
tuna = data[data['species'].str.contains('tuna', case=False)]
tuna
```

```python
tuna.sort_values('F_Fmsy', inplace=True)
```

```python
fig, ax = plt.subplots(figsize=(20, 10))

for i, row in tuna.iterrows():
    ax.scatter(row['B_Bmsy'], row['F_Fmsy'], label=row['species'] + ' (' + str(row['area_code']) + ')',s=150 , marker="o")

ax.axhline(1, color='black', linestyle='--', label='F_Fmsy = 1')
ax.axvline(1, color='black', linestyle='--', label='B_Bmsy = 1')

ax.fill_between([1, 2],1,2,alpha=0.3, color='#F38D25')  # orange
ax.fill_between([0, 1], 0, 1, alpha=0.3, color='#F38D25')  # orange
ax.fill_between([1, 0], 1, 2, alpha=0.3, color='#DA383D')  # red
ax.fill_between([1, 2], 0, 1, alpha=0.3, color='#1F98D0')  # blue

ax.set_xlabel('B_Bmsy')
ax.set_ylabel('F_Fmsy')
ax.legend(loc='upper left', bbox_to_anchor=(1, 1))
ax.set_title('Kobe Plot - Tuna Species')

plt.show()
```
